```python
# Empty lists for each category
low = []        # Less than 3 hours
moderate = []   # Between 3 and 5 hours
high = []       # More than 5 hours
```

```python
# Example study hours list
time_data = [2, 3.5, 6, 1, 4, 5.5, 2.8]

# Empty lists for categories
low = []        # < 3 hours
moderate = []   # 3 to 5 hours
high = []       # > 5 hours

# Iterate and classify
for time in time_data:
    if time < 3:
        low.append(time)
    elif 3 <= time <= 5:
        moderate.append(time)
    else:  # time > 5
        high.append(time)

# Print results
print("Low:", low)
print("Moderate:", moderate)
print("High:", high)
```

```python
# Example daily records: [study, entertainment, sleep]
time_data = [
    [2, 3, 7],
    [3.5, 2, 6.5],
    [6, 1, 8],
    [1, 4, 6],
    [4, 2, 7],
    [5.5, 3, 6],
    [2.8, 2.5, 7.2]
]

# Step 1: Create empty lists
study_hours = []
entertainment_hours = []
sleep_hours = []

# Step 2: Extract values into each list
for record in time_data:
    study_hours.append(record[0])
    entertainment_hours.append(record[1])
    sleep_hours.append(record[2])

# Step 3: Calculate and print averages
average_study = sum(study_hours) / len(study_hours)
average_entertainment = sum(entertainment_hours) / len(entertainment_hours)
average_sleep = sum(sleep_hours) / len(sleep_hours)

print("Average hours spent studying:", average_study)
print("Average hours spent on entertainment:", average_entertainment)
print("Average hours spent sleeping:", average_sleep)
```

```python
def sum_nested_list(nested_list):
    total = 0
    for item in nested_list:
        # Case 1: If the item is a list, recurse
        if isinstance(item, list):
            total += sum_nested_list(item)
        # Case 2: If the item is a number, add it directly
        else:
            total += item
    return total

```

```python
data = [1, [2, 3], [4, [5, 6]]]
print(sum_nested_list(data))   # Output: 21
```

```python
def calculate_directory_size(directory):
    total_size = 0

    for key, value in directory.items():
        # Case 1: If the value is a nested directory (another dictionary)
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        # Case 2: If the value is a file size (number)
        else:
            total_size += value

    return total_size
```

```python
def min_coins(coins, amount):
    # Create DP array where dp[i] = minimum coins to make amount i
    dp = [float('inf')] * (amount + 1)
    dp[0] = 1  # Base case: 0 coins needed to form amount 0

    # Fill up dp array
    for a in range(1, amount + 1):
        for coin in coins:
            if a - coin >= 0:
                dp[a] = min(dp[a], dp[a - coin] + 1)

    # If dp[amount] is still infinity, solution doesn't exist
    return dp[amount] if dp[amount] != float('inf') else -1
```
