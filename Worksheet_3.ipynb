Q1: Who paid the highest fare per year of age?
Fill missing Age with median

```python
#Fill missing Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Add fare_per_year
df['fare_per_year'] = df['Fare'] / df['Age']

#Subset fare_per_year > 5
high_fare_age = df[df['fare_per_year'] > 5]

#Sort descending
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

#Select Name + fare_per_year
result = high_fare_age_srt[['Name', 'fare_per_year']]
print(result.head())
```

Q2: Which adult male paid highest fare relative to class?

```python
#Add fare_per_class
df['fare_per_class'] = df['Fare'] / df['Pclass']

#Subset adult males
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]

#Sort descending
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

#Select Name, Age, fare_per_class
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]
print(result.head())
```

3.3 – GROUP-BY PRACTICE EXERCISE
Q1: % of total fare revenue per class

```python
#Total fare
total_fare = df['Fare'].sum()

#Totals by class
fare_1 = df[df['Pclass'] == 1]['Fare'].sum()
fare_2 = df[df['Pclass'] == 2]['Fare'].sum()
fare_3 = df[df['Pclass'] == 3]['Fare'].sum()

#Combine
fare_list = [fare_1, fare_2, fare_3]

#Percent contribution
percent = [(x / total_fare) * 100 for x in fare_list]
print({
    'Class 1 %': percent[0],
    'Class 2 %': percent[1],
    'Class 3 %': percent[2]
})
```

Q2: % of passengers per age group (child/adult/senior)

```python
#Create age_group column
def group_age(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

df['age_group'] = df['Age'].apply(group_age)

#Total passengers
total_passengers = len(df)

#Count each age group
group_counts = df['age_group'].value_counts()

#Proportion
percentages = (group_counts / total_passengers) * 100

#Display
print(percentages)
```
